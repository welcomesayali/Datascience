Que 1: Name any five plots using Seaborn and their uses
1. Line Plot (sns.lineplot)

Used to show trends over time or a continuous variable.

Example: Plotting temperature over days.

2. Box Plot (sns.boxplot)

Displays distribution and outliers in numerical data grouped by categories.

Example: Comparing ages across different classes on the Titanic.

3. Histogram (sns.histplot)

Shows the distribution of a single numerical variable.

Example: Frequency of different price ranges in the diamonds dataset.

4. Scatter Plot (sns.scatterplot)

Shows the relationship between two numeric variables.

Example: Relationship between sepal length and width in the iris dataset.

5. Heatmap (sns.heatmap)

Displays matrix-style data, commonly used with correlations or frequency tables.

Example: Passenger traffic per month/year in the flights dataset.

Q2: Line plot with fmri dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
fmri = sns.load_dataset("fmri")

# Line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Signal over Time by Event and Region")
plt.show()
```

Q3: Box plots with titanic dataset

```python
# Load dataset
titanic = sns.load_dataset("titanic")

# Box plot for age
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title("Passenger Age by Class")
plt.show()

# Box plot for fare
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title("Passenger Fare by Class")
plt.show()
```

Q4: Histogram with diamonds dataset

```python
# Load dataset
diamonds = sns.load_dataset("diamonds")

# Histogram for price with hue based on cut
sns.histplot(data=diamonds, x='price', hue='cut', kde=True, multiple='stack')
plt.title("Price Distribution by Cut Quality")
plt.show()
```

Q5: Pair plot with iris dataset

```python
# Load dataset
iris = sns.load_dataset("iris")

# Pair plot
sns.pairplot(iris, hue="species")
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

Q6: Heatmap with flights dataset

```python
# Load dataset
flights = sns.load_dataset("flights")

# Pivot the data for heatmap
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

# Heatmap
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Monthly Passenger Count (Heatmap)")
plt.show()
```
