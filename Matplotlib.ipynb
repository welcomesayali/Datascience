Q1: What is Matplotlib? Why is it used?
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

It is widely used in data visualization to understand data patterns and trends.

The pyplot module in Matplotlib offers a simple interface similar to MATLAB.

Five common plot types in matplotlib.pyplot:

Line plot

Bar plot

Scatter plot

Histogram

Box plot

Q2: What is a scatter plot?
A scatter plot displays values for two variables as points on a two-dimensional graph, useful for spotting correlations or trends.


```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

Q3: Why is subplot() used?
subplot() is used to display multiple plots in one figure by dividing the figure into a grid layout.

```python
#Code to draw 4 line plots using subplot():

import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("Line 4")

plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used?
A bar plot displays categorical data with rectangular bars. It is used to compare quantities across different categories.

```python
# Code to create a bar plot and horizontal bar plot:

import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.bar(company, profit)
plt.title("Company Profit (Bar Plot)")
plt.ylabel("Profit")
plt.show()

# Horizontal Bar Plot
plt.barh(company, profit)
plt.title("Company Profit (Horizontal Bar Plot)")
plt.xlabel("Profit")
plt.show()
```

Q5: What is a box plot? Why is it used?
A box plot (or box-and-whisker plot) shows the distribution of a dataset by displaying:

Minimum

First quartile (Q1)

Median (Q2)

Third quartile (Q3)

Maximum

Outliers (if any)

It's used to visualize spread and detect outliers in numerical data.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
plt.boxplot(data, labels=["Box1", "Box2"])
plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show()
```
