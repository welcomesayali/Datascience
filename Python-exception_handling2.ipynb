Q1. Why do we have to use the Exception class while creating a Custom Exception?
We inherit from the built-in Exception class when creating a custom exception because:

Exception is the base class for all built-in exceptions.

It provides a consistent interface and behavior for exception handling.

Inheriting from Exception allows custom exceptions to be:

Raised using raise

Caught using except

Recognized by built-in error tools and debuggers

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
The ArithmeticError class is a base class for errors occurring in numeric calculations.

Errors under ArithmeticError:

ZeroDivisionError

OverflowError

FloatingPointError

```python
print(10 / 0)  # Raises ZeroDivisionError
import math
print(math.exp(1000))  # Raises OverflowError
```

Q4. Why is the LookupError class used? Explain with KeyError and IndexError.
LookupError is the base class for all lookup-related errors.

It's used when trying to access an invalid index or key in sequences like lists, dictionaries, etc.

```python
my_dict = {"a": 1}
print(my_dict["b"])  # Raises KeyError
```

```python
my_list = [1, 2, 3]
print(my_list[5])  # Raises IndexError
```
